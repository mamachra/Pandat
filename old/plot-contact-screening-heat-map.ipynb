```python
import csv
s = open('/Users/mahdiamachraa1/Desktop/angle-screening.csv')
s_Reader = csv.reader(s)
data = list(s_Reader)

theta1 =  [float(data[i][1]) for i in range(1,len(data))]
theta2 = [float(data[i][2]) for i in range(1,len(data))]
T = [float((data[i][3])) for i in range(1,len(data))]
```

```python
from pymatgen.util.plotting import get_publication_quality_plot
from pymatgen.util.plotting import pretty_plot
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt = get_publication_quality_plot(12,8, plt=plt)
```

```python
from pymatgen.util.plotting import pretty_plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt = pretty_plot(12,8, plt=plt)

data = pd.DataFrame({'Ag-angle': theta1, 'Sn-angle': theta2, 'T': T})
data_pivoted = data.pivot("Ag-angle", "Sn-angle", "T")
plt = sns.heatmap(data_pivoted, linewidth = 0.2 ,vmin = 203, vmax= 264,cmap="YlGnBu",annot=True)
```
